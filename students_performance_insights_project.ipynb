```python
# ---------------- Student Performance Insights ---------------- #
import pandas as pd
import matplotlib.pyplot as plt

# --- 1. Create Sample Dataset ---
student_data = {
    'Student': ['Aisha','Rahul','Sneha','Arjun','Priya','Kiran','Meena'],
    'Subject': ['Math','English','Science','Math','Science','English','Math'],
    'Marks': [87,75,92,68,89,80,95]
}

df = pd.DataFrame(student_data)
df.to_csv('student_scores.csv', index=False)
print("✅ student_scores.csv file created\n")

# --- 2. Load Data ---
data = pd.read_csv('student_scores.csv')
print(data.head(), "\nAverage Marks:", data['Marks'].mean())

# --- 3. Subject-wise Performance ---
subject_perf = data.groupby('Subject')['Marks'].mean().sort_values(ascending=False)
print("\nAverage Marks per Subject:\n", subject_perf)

# --- 4. Visualization ---
plt.figure(figsize=(8,5))
subject_perf.plot(kind='bar', color='orange')
plt.title('Average Marks by Subject')
plt.xlabel('Subject'); plt.ylabel('Average Score'); plt.tight_layout(); plt.show()

# --- 5. Top Students ---
print("\nTop 5 Students:\n", data.sort_values(by='Marks', ascending=False).head(5))
```
